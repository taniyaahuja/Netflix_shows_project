# Data Cleaning

**Setting Show ID as the index**

```python
a.set_index("show_id", inplace=True)
```

**Checking for null values**

```python
mask1= a.isnull().sum()
mask1
```

**Filling null values with "NaN"**

```python
a["director"]= a["director"].fillna("NaN")
a["cast"]= a["cast"].fillna("NaN")
a["country"]= a["country"].fillna("NaN")
a["rating"]= a["rating"].fillna("NaN")
a["duration"]= a["duration"].fillna("NaN")
```

**Checking for unique elements in "type" column**

```python
a["type"].unique()
```

**Changing data type**

```python
a["type"]= a["type"].astype("category")
a['release_year'] = pd.to_datetime(a['release_year'], errors='coerce', format='%Y')
a['date_added'] = pd.to_datetime(a['date_added'], errors='coerce', format='%B %d, %Y').fillna(pd.Timestamp('1970-01-01'))     #here we have filled the null values with a default date
```

**Changing column headers to proper case**

```python
a.columns= a.columns.str.title()
```

# 1. Show Types Distribution

```python
ct= a["Type"].value_counts()
ct
```

```python
ct.plot(kind= "bar", color= ["lightgreen","lightblue"], edgecolor= "black")
plt.title("Show Types Distribution")
plt.ylabel("Frequency")
plt.show()
```

Findings in (1):
* The count of Movies is more than twice the count of TV shows.

# 2. Top 10 Ratings on Netflix

```python
c = a["Rating"].value_counts().head(10)
plt.figure(figsize=(10,10))
plt.pie(c, autopct="%.1f%%", labels= c.index)
plt.title("Top 10 Ratings on Netflix")
plt.show()
```

Findings in (2):
* "TV-MA" & "TV-14" dominate the ratings pie chart with more than 60% of the total count combined.

# 3. Release Trends Over Years (2002-2021)

```python
filtered_data = a[(a["Release_Year"].dt.year>=2002) & (a["Release_Year"].dt.year<=2021)]
a["Release_Year"] = a["Release_Year"].dt.year
cby = filtered_data["Release_Year"].dt.year.value_counts().sort_index()

filtered_data2 = a[(a['Date_Added'].dt.year >= 2002) & (a['Date_Added'].dt.year <= 2021)]
a['Date_Added'] = a['Date_Added'].dt.year
cby2 = filtered_data2['Date_Added'].dt.year.value_counts().sort_index()
```

```python
plt.figure(figsize=(20, 6))
plt.plot(cby.index, cby.values, marker='o', label="Released shows")
plt.plot(cby2.index, cby2.values, marker='o', linestyle='dashed', label="Added shows")
plt.xticks([2002,2004,2006,2008,2010,2012,2014,2016,2018,2020,2022])
plt.title("Number of shows released (2002-2021) vs Number of shows added on netflix (2008-2021)")
plt.ylabel("Number of shows")
plt.xlabel("Year")
plt.legend()
plt.grid()
plt.show()
```

Finding  in (3):
* The number of shows released(blue trend) spiked from 2014 but declined sharply in 2021
* The number of shows added(orange trend) spiked from 2015 but declined sharply in 2021

# 4. Top Country of origin of shows

```python
country = a["Country"].value_counts().head(15)
plt.figure(figsize=(10,10))
plt.pie(country, autopct="%.1f%%", labels= country.index)
plt.title("Top 15 show's Country of origin")
plt.show()
```

Findings in (4):
* More than 40% of the shows were from United States while almost 15% were from India.
* "NaN" is for the null values in the dataset, ie. the country is not specified in the dataset
